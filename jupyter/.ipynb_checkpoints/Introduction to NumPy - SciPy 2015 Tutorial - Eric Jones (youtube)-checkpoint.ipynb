```python
x = np.linspace(0, 2 * np.pi, 50)
plt.plot(np.sin(x), label = 'Plot A')
plt.xlabel('Item #')
plt.ylabel('Sin(Item)')
plt.grid()
plt.legend()
# plt.hold(True)
# plt.clf()
# plt.close()
# plt.close('all')
```

```python
plt.plot(x, np.sin(x), 'g-.', x, np.sin(2 * x), 'b.')
```

```python
x1 = np.random.rand(200)
y1 = np.random.rand(200)

size = np.random.rand(200) * 30
color = np.random.rand(200)

plt.scatter(x1, y1, size, color)
plt.colorbar()
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x), 'g-.', x, np.sin(2 * x), 'b.')
plt.legend(['Plot A', 'Plot B'])
plt.subplot(2, 1, 2)
plt.scatter(x1, y1, size, color)
```

```python
from scipy.misc import ascent
img = ascent()
plt.imshow(img, cmap = cm.viridis)
```

```python


img = face()
plt.imshow(img, extent=[-50, 50, -50, 50], cmap = cm.viridis)
plt.colorbar()
```

```python
plt.matshow(np.random.rand(64, 64), fignum=100, cmap=plt.cm.viridis)

plt.show()
```

```python
m = np.array([[1, 2, 3], [4, 5, 6]])
plt.subplot(2, 1, 1)
plt.matshow(m, cmap=plt.cm.viridis)
```

```python
grid = np.random.rand(10, 10)
plt.subplot(3, 1, 1)
plt.imshow(grid, cmap = cm.viridis)

grid2 = grid.copy()
grid2_c = grid2[1:-1, 1:-1]
grid2_u = grid2[:-2, 1:-1]
grid2_a = (grid2_c + grid2_u)/2

plt.subplot(3, 1, 2)
plt.imshow(grid2_a, cmap = cm.viridis)


plt.subplot(3, 1, 3)
plt.imshow(grid2, interpolation='gaussian', cmap = cm.viridis)
```

## Fancy Indexing

```python
a = np.arange(0, 2 * np.pi, 0.1, dtype=float)
a
```

```python
index = [3, 7, 8]
y = a[index]
y
```

```python
y[0] = -1
a
```

```python
x = a[1:5]
x[0] = -1
a
```

```python
a > 50
```

```python
b = np.sin(a)
```

```python
plt.plot(a, b)
mask = b > 0
plt.plot(a[mask], b[mask], 'r.')
```

```python
a = np.zeros(100)
a.shape = 10, 10

rows, cols = a.shape
a[np.arange(rows - 1), np.arange(1, cols)] = 10

plt.imshow(a, cmap = cm.viridis)
plt.colorbar()
```

```python
plt.imshow(a == 0)
```
